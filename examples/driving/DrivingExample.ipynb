# Train a Machine Learning Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

Use Pandas to divide the data into a 'feature' maxtrix and a 'target' matrix.

Remember - column 0 is the target values of left, straight, right.  Columns 1 to the end represent the road configuation.

Using some Pandas syntax:

.loc - this will locate data in the dataframe.

The syntax is:  .loc[rows, columns].

Using ':' is short hand for 'all of the values'.

so `df.loc[:, 1:]`, means all of the rows, and columns 1 to end.

```python
X = df.loc[:, 1:]
y = df.loc[:, 0]
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, shuffle=True)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_train = model.predict(X_train)
```

```python
accuracy_score(y_train, y_pred_train)
```

# Test Machine Learning Model

```python
y_pred_test = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_test)
```

Before we make new predictions - lets re-train the model on all of the data so the model has a chance to learn from all of the driving data.

```python
model = RandomForestClassifier()
model.fit(X, y)
```

```python
# Save the model so we can use it to autonmously drive the car
import joblib

joblib.dump(model, "./scripts/best_driving_model.sav")
```
